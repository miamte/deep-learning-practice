# 5мһҘ. м„ нҳ• нҡҢк·Җ лӘЁлҚё: лЁјм Җ кёӢкі  мҲҳм •н•ҳкё°

## 2. нҢҢмқҙмҚ¬ мҪ”л”©мңјлЎң нҷ•мқён•ҳлҠ” м„ нҳ• нҡҢк·Җ

### 1.нҷҳкІҪ мӨҖл№„

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2. лҚ°мқҙн„° мӨҖл№„

```python
# кіөл¶Җ мӢңк°„ XмҷҖ м„ұм Ғ yмқҳ л„ҳнҢҢмқҙ л°°м—ҙмқ„ л§Ңл“ӯлӢҲлӢӨ.
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

### 3. лҚ°мқҙн„° л¶„нҸ¬ нҷ•мқё

```python
# лҚ°мқҙн„°мқҳ л¶„нҸ¬лҘј к·ёлһҳн”„лЎң лӮҳнғҖлғ…лӢҲлӢӨ.
plt.scatter(x, y)
plt.show()
```

### 4. мӢӨн–үмқ„ мң„н•ң ліҖмҲҳ м„Өм •

```python
# кё°мҡёкё° aмҷҖ м ҲнҺё bмқҳ к°’мқ„ мҙҲкё°нҷ”н•©лӢҲлӢӨ.
a = 0
b = 0

# н•ҷмҠөлҘ мқ„ м •н•©лӢҲлӢӨ.
lr = 0.03

# лӘҮ лІҲ л°ҳліөлҗ м§Җ м„Өм •н•©лӢҲлӢӨ.
epochs = 2001
```

### 5. кІҪмӮ¬ н•ҳк°•лІ•

```python
# x к°’мқҙ мҙқ лӘҮ к°ңмқём§Җ м…үлӢҲлӢӨ.
n=len(x)

# кІҪмӮ¬ н•ҳк°•лІ•мқ„ мӢңмһ‘н•©лӢҲлӢӨ.
for i in range(epochs):                  # м—җнҸ¬нҒ¬ мҲҳ л§ҢнҒј л°ҳліө

    y_pred = a * x + b                   # мҳҲмёЎ к°’мқ„ кө¬н•ҳлҠ” мӢқмһ…лӢҲлӢӨ.
    error = y - y_pred                   # мӢӨм ң к°’кіј л№„көҗн•ң мҳӨм°ЁлҘј errorлЎң лҶ“мҠөлӢҲлӢӨ.

    a_diff = (2/n) * sum(-x * (error))   # мҳӨм°Ё н•ЁмҲҳлҘј aлЎң нҺёлҜёл¶„н•ң к°’мһ…лӢҲлӢӨ.
    b_diff = (2/n) * sum(-(error))       # мҳӨм°Ё н•ЁмҲҳлҘј bлЎң нҺёлҜёл¶„н•ң к°’мһ…лӢҲлӢӨ.

    a = a - lr * a_diff     # н•ҷмҠөлҘ мқ„ кіұн•ҙ кё°мЎҙмқҳ a к°’мқ„ м—…лҚ°мқҙнҠён•©лӢҲлӢӨ.
    b = b - lr * b_diff     # н•ҷмҠөлҘ мқ„ кіұн•ҙ кё°мЎҙмқҳ b к°’мқ„ м—…лҚ°мқҙнҠён•©лӢҲлӢӨ.

    if i % 100 == 0:        # 100лІҲ л°ҳліөлҗ  л•Ңл§ҲлӢӨ нҳ„мһ¬мқҳ a к°’, b к°’мқ„ м¶ңл Ҙн•©лӢҲлӢӨ.
        print("epoch=%.f, кё°мҡёкё°=%.04f, м ҲнҺё=%.04f" % (i, a, b))
```

### 6. к·ёлһҳн”„ нҷ•мқё

```python
# м•һм„ң кө¬н•ң мөңмў… a к°’мқ„ кё°мҡёкё°, b к°’мқ„ y м ҲнҺём—җ лҢҖмһ…н•ҳм—¬ к·ёлһҳн”„лҘј к·ёлҰҪлӢҲлӢӨ.
y_pred = a * x + b

# к·ёлһҳн”„ м¶ңл Ҙ
plt.scatter(x, y)
plt.plot(x, y_pred,'r')
plt.show()
```

## 4. нҢҢмқҙмҚ¬ мҪ”л”©мңјлЎң нҷ•мқён•ҳлҠ” лӢӨмӨ‘ м„ нҳ• нҡҢк·Җ

### 1.нҷҳкІҪ мӨҖл№„

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2. лҚ°мқҙн„° мӨҖл№„

```python
# кіөл¶Җ мӢңк°„ x1кіј кіјмҷё мӢңк°„ x2, к·ё м„ұм Ғ yмқҳ л„ҳнҢҢмқҙ л°°м—ҙмқ„ л§Ңл“ӯлӢҲлӢӨ.
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])
```

### 3. лҚ°мқҙн„° л¶„нҸ¬ нҷ•мқё

```python
# лҚ°мқҙн„°мқҳ л¶„нҸ¬лҘј к·ёлһҳн”„лЎң лӮҳнғҖлғ…лӢҲлӢӨ.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
plt.show()
```

### 4. мӢӨн–үмқ„ мң„н•ң ліҖмҲҳ м„Өм •

```python
# кё°мҡёкё° aмҷҖ м ҲнҺё bмқҳ к°’мқ„ мҙҲкё°нҷ”н•©лӢҲлӢӨ.
a1 = 0
a2 = 0
b = 0

# н•ҷмҠөлҘ мқ„ м •н•©лӢҲлӢӨ.
lr = 0.01

# лӘҮ лІҲ л°ҳліөлҗ м§Җ м„Өм •н•©лӢҲлӢӨ.
epochs = 2001
```

### 5. кІҪмӮ¬ н•ҳк°•лІ•

```python
# x к°’мқҙ мҙқ лӘҮ к°ңмқём§Җ м…үлӢҲлӢӨ. x1кіј x2мқҳ мҲҳк°Җ к°ҷмңјлҜҖлЎң x1л§Ң м„ёкІ мҠөлӢҲлӢӨ.
n=len(x1)

# кІҪмӮ¬ н•ҳк°•лІ•мқ„ мӢңмһ‘н•©лӢҲлӢӨ.
for i in range(epochs):                  # м—җнҸ¬нҒ¬ мҲҳ л§ҢнҒј л°ҳліө

    y_pred = a1 * x1 + a2 * x2 + b       # мҳҲмёЎ к°’мқ„ кө¬н•ҳлҠ” мӢқмқ„ м„ёмӣҒлӢҲлӢӨ
    error = y - y_pred                   # мӢӨм ң к°’кіј л№„көҗн•ң мҳӨм°ЁлҘј errorлЎң лҶ“мҠөлӢҲлӢӨ.

    a1_diff = (2/n) * sum(-x1 * (error)) # мҳӨм°Ё н•ЁмҲҳлҘј a1лЎң нҺёлҜёл¶„н•ң к°’мһ…лӢҲлӢӨ.
    a2_diff = (2/n) * sum(-x2 * (error)) # мҳӨм°Ё н•ЁмҲҳлҘј a2лЎң нҺёлҜёл¶„н•ң к°’мһ…лӢҲлӢӨ.
    b_diff = (2/n) * sum(-(error))       # мҳӨм°Ё н•ЁмҲҳлҘј bлЎң нҺёлҜёл¶„н•ң к°’мһ…лӢҲлӢӨ.

    a1 = a1 - lr * a1_diff  # н•ҷмҠөлҘ мқ„ кіұн•ҙ кё°мЎҙмқҳ a1 к°’мқ„ м—…лҚ°мқҙнҠён•©лӢҲлӢӨ.
    a2 = a2 - lr * a2_diff  # н•ҷмҠөлҘ мқ„ кіұн•ҙ кё°мЎҙмқҳ a2 к°’мқ„ м—…лҚ°мқҙнҠён•©лӢҲлӢӨ.
    b = b - lr * b_diff     # н•ҷмҠөлҘ мқ„ кіұн•ҙ кё°мЎҙмқҳ b к°’мқ„ м—…лҚ°мқҙнҠён•©лӢҲлӢӨ.

    if i % 100 == 0:        # 100лІҲ л°ҳліөлҗ  л•Ңл§ҲлӢӨ нҳ„мһ¬мқҳ a1, a2, b к°’мқ„ м¶ңл Ҙн•©лӢҲлӢӨ.
        print("epoch=%.f, кё°мҡёкё°1=%.04f, кё°мҡёкё°2=%.04f, м ҲнҺё=%.04f" % (i, a1, a2, b))
```

```python
# мӢӨм ң м җмҲҳмҷҖ мҳҲмёЎлҗң м җмҲҳлҘј м¶ңл Ҙн•©лӢҲлӢӨ.
print("мӢӨм ң м җмҲҳ:", y)
print("мҳҲмёЎ м җмҲҳ:", y_pred)
```

## 5-1. н…җм„ңн”ҢлЎңм—җм„ң мӢӨн–үн•ҳлҠ” м„ нҳ• нҡҢк·Җ лӘЁлҚё

### 1. нҷҳкІҪ мӨҖл№„

```python
import numpy as np
import matplotlib.pyplot as plt

# н…җм„ңн”ҢлЎңмқҳ мјҖлқјмҠӨ APIм—җм„ң н•„мҡ”н•ң н•ЁмҲҳл“Өмқ„ л¶Ҳлҹ¬мҳөлӢҲлӢӨ.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2. лҚ°мқҙн„° мӨҖл№„

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

### 3. лӘЁлҚё мӢӨн–ү

```python
model = Sequential()

# м¶ңл Ҙ к°’, мһ…л Ҙ ліҖмҲҳ, л¶„м„қ л°©лІ•м—җ л§һкІҢлҒ” лӘЁлҚёмқ„ м„Өм •н•©лӢҲлӢӨ.
model.add(Dense(1, input_dim=1, activation='linear'))

# мҳӨм°Ё мҲҳм •мқ„ мң„н•ҙ кІҪмӮ¬ н•ҳк°•лІ•(sgd)мқ„, мҳӨм°Ёмқҳ м •лҸ„лҘј нҢҗлӢЁн•ҳкё° мң„н•ҙ нҸүк·  м ңкіұ мҳӨм°Ё(mse)лҘј мӮ¬мҡ©н•©лӢҲлӢӨ.
model.compile(optimizer='sgd', loss='mse')

# мҳӨм°ЁлҘј мөңмҶҢнҷ”н•ҳлҠ” кіјм •мқ„ 2000лІҲ л°ҳліөн•©лӢҲлӢӨ.
model.fit(x, y, epochs=2000)
```

### 4. к·ёлһҳн”„лЎң нҷ•мқё

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x),'r')    # мҳҲмёЎ кІ°кіјлҘј к·ёлһҳн”„лЎң лӮҳнғҖлғ…лӢҲлӢӨ.
plt.show()
```

### 5. лӘЁлҚё н…ҢмҠӨнҠё

```python
import tensorflow as tf

# мһ„мқҳмқҳ мӢңк°„мқ„ м§‘м–ҙл„Јм–ҙ м җмҲҳлҘј мҳҲмёЎн•ҳлҠ” лӘЁлҚёмқ„ н…ҢмҠӨнҠён•ҙ ліҙкІ мҠөлӢҲлӢӨ.

hour = 7
#prediction = model.predict([hour]) #н…җм„ңн”ҢлЎң 2.9.0 к№Ңм§Җл§Ң к°ҖлҠҘ

input_data = tf.constant([[hour]]) # н…җм„ңн”ҢлЎң лІ„м „м—…мңјлЎң мқён•ң мҲҳм •
prediction = model.predict(input_data)[0][0]  # н…җм„ңн”ҢлЎң лІ„м „м—…мңјлЎң мқён•ң мҲҳм •

print("%.fмӢңк°„мқ„ кіөл¶Җн•  кІҪмҡ°мқҳ мҳҲмғҒ м җмҲҳлҠ” %.02fм җмһ…лӢҲлӢӨ" % (hour, prediction))
```

## 5-2 н…җм„ңн”ҢлЎңм—җм„ң мӢӨн–үн•ҳлҠ” лӢӨмӨ‘ м„ нҳ• нҡҢк·Җ лӘЁлҚё

### 1.нҷҳкІҪ мӨҖл№„

```python
import numpy as np
import matplotlib.pyplot as plt

# н…җм„ңн”ҢлЎңмқҳ мјҖлқјмҠӨ APIм—җм„ң н•„мҡ”н•ң н•ЁмҲҳл“Өмқ„ л¶Ҳлҹ¬мҳөлӢҲлӢӨ.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2. лҚ°мқҙн„° мӨҖл№„

```python
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97])
```

### 3. лӘЁлҚё мӢӨн–ү

```python
model = Sequential()

# мһ…л Ҙ ліҖмҲҳк°Җ 2к°ң(н•ҷмҠө мӢңк°„, кіјмҷё мӢңк°„)мқҙлҜҖлЎң input_dimм—җ 2лҘј мһ…л Ҙн•©лӢҲлӢӨ.
model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd' ,loss='mse')

model.fit(x, y, epochs=20)
```

### 4. лӘЁлҚё н…ҢмҠӨнҠё

```python
import tensorflow as tf

# мһ„мқҳмқҳ н•ҷмҠө мӢңк°„кіј кіјмҷё мӢңк°„мқ„ м§‘м–ҙл„Јм–ҙ м җмҲҳлҘј мҳҲмёЎн•ҳлҠ” лӘЁлҚёмқ„ н…ҢмҠӨнҠён•ҙ ліҙкІ мҠөлӢҲлӢӨ.

hour = 7
private_class = 4
#prediction = model.predict([[hour, private_class]])  #н…җм„ңн”ҢлЎң 2.9.0 к№Ңм§Җл§Ң к°ҖлҠҘ

input_data = tf.constant([[hour, private_class]])  # н…җм„ңн”ҢлЎң лІ„м „м—…мңјлЎң мқён•ң мҲҳм •
prediction = model.predict(input_data)[0][0]  # н…җм„ңн”ҢлЎң лІ„м „м—…мңјлЎң мқён•ң мҲҳм •

print("%.fмӢңк°„мқ„ кіөл¶Җн•ҳкі  %.fмӢңк°„мқҳ кіјмҷёлҘј л°ӣмқ„ кІҪмҡ°, мҳҲмғҒ м җмҲҳлҠ” %.02fм җмһ…лӢҲлӢӨ" % (hour, private_class, prediction))
```
