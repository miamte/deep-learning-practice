# 코드를 통해서 XOR 문제 해결하기

```python
# XOR 문제를 해결하기 위한 가중치와 바이어스 값을 정의합니다.
w11 = np.array([-2, -2]) # 첫 번째 은닉층 뉴런의 가중치
w12 = np.array([2, 2])   # 두 번째 은닉층 뉴런의 가중치
w2 = np.array([1, 1])    # 출력층 뉴런의 가중치

b1 = 3 # 첫 번째 은닉층 뉴런의 바이어스
b2 = -1 # 두 번째 은닉층 뉴런의 바이어스
b3 = -1 # 출력층 뉴런의 바이어스
```

```python
# 퍼셉트론 함수를 정의합니다.
# 입력 x, 가중치 w, 바이어스 b를 받아 활성화 함수를 적용한 결과를 반환합니다.
def MLP(x, w, b):
    # 가중치와 입력의 곱의 합에 바이어스를 더합니다.
    y = np.sum(w * x) + b

    # 활성화 함수 (계단 함수): y가 0보다 작거나 같으면 0, 그렇지 않으면 1을 반환합니다.
    if y <= 0:
        return 0
    else:
        return 1
```

```python
# NAND 게이트 함수를 정의합니다. MLP 함수와 w11, b1을 사용합니다.
def NAND(x1,x2):
    return MLP(np.array([x1, x2]), w11, b1)

# OR 게이트 함수를 정의합니다. MLP 함수와 w12, b2를 사용합니다.
def OR(x1,x2):
    return MLP(np.array([x1, x2]), w12, b2)

# AND 게이트 함수를 정의합니다. MLP 함수와 w2, b3를 사용합니다.
def AND(x1,x2):
    return MLP(np.array([x1, x2]), w2, b3)

# XOR 게이트 함수를 정의합니다. NAND, OR, AND 게이트를 조합하여 구현합니다.
def XOR(x1,x2):
    # NAND 게이트와 OR 게이트의 출력을 AND 게이트의 입력으로 사용합니다.
    return AND(NAND(x1, x2),OR(x1,x2))
```

```python
# XOR 게이트의 모든 가능한 입력 값에 대해 최종 값을 출력합니다.
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    # XOR 함수를 호출하여 결과를 얻습니다.
    y = XOR(x[0], x[1])
    # 입력 값과 출력 값을 출력합니다.
    print("입력 값: " + str(x) + " 출력 값: " + str(y))
```

# Multi Perceptron Class 생성을 통해 해결하기

```python
# XOR 문제를 해결하기 위한 다층 퍼셉트론(Multi-Layer Perceptron) 클래스를 정의합니다.
class SimpleMLP:
    def __init__(self):
        # 수동으로 설정한 가중치와 바이어스
        # 입력층에서 은닉층으로 가는 가중치 (x1, x2 -> n1, n2 계산)
        self.weights_input_hidden = np.array([[-2, 2], [-2, 2]])
        # 은닉층 뉴런들의 바이어스 (x1, x2 -> n1, n2)
        self.bias_hidden = np.array([3, -1])

        # 출력층: 1개의 뉴런
        # 은닉층에서 출력층으로 가는 가중치 (n1, n2 -> y)
        self.weights_hidden_output = np.array([1, 1])
        # 출력층 뉴런의 바이어스 (n1, n2 -> y)
        self.bias_output = -1

    # 순전파(forward) 메서드를 정의합니다.
    def forward(self, X):
        # 은닉층 계산:
        # 입력 X와 입력-은닉층 가중치를 내적하고 은닉층 바이어스를 더합니다.
        hidden_input = np.dot(X, self.weights_input_hidden) + self.bias_hidden
        # 은닉층의 활성화 함수 (계단 함수): 0보다 작거나 같으면 0, 아니면 1
        hidden_output = np.where(hidden_input <= 0, 0, 1)
        print(f"hidden_output : {hidden_input}")

        # 출력층 계산:
        # 은닉층 출력과 은닉-출력층 가중치를 내적하고 출력층 바이어스를 더합니다.
        final_input = np.dot(hidden_output, self.weights_hidden_output) + self.bias_output
        # 출력층의 활성화 함수 (계단 함수): 0보다 작거나 같으면 0, 아니면 1
        final_output = np.where(final_input <= 0, 0, 1)
        print(f"final_output : {final_input}")

        # 최종 출력을 반환합니다.
        return final_output
```

```python
# XOR 문제의 입력값(X)과 실제 출력값(y) 데이터셋을 정의합니다.
XOR_X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # 입력값
XOR_y = np.array([0, 1, 1, 0])  # XOR 출력값
```

```python
# SimpleMLP 모델을 초기화합니다.
mlp = SimpleMLP()
```

```python
# 정의된 XOR 입력값(XOR_X)에 대해 모델의 순전파를 실행하여 예측값을 얻습니다.
predictions = mlp.forward(XOR_X)
# 예측 결과를 출력합니다.
print(f"XOR 문제 예측 결과: {predictions}")
```
